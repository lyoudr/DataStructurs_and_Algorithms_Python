```python
from typing import List 
# 347. Top K Frequent Elements
"""
Given an integer array nums and an integer k, 
return the k most frequent elements. 
You may return the answer in any order.
"""
class MaxHeap:
    def __init__(self):
        self.heap = []
    
    def insert(self, value):
        self.heap.append(value)
        self._heapify_up()
    
    def extract_max(self):
        if len(self.heap) == 0:
            return []

        # Swap the root with the last element 
        self.heap[0], self.heap[-1] = self.heap[-1], self.heap[0]
        # Extract the maximum value
        max_value = self.heap.pop()
        self._heapify_down()
        return max_value[1]
    
    def _heapify_up(self):
        index = len(self.heap) - 1
        while index > 0:
            parent_index = (index - 1) // 2
            if self.heap[index] > self.heap[parent_index]:
                self.heap[index], self.heap[parent_index] = self.heap[parent_index], self.heap[index]
                index = parent_index 
            else:
                break 
    
    def _heapify_down(self):
        index = 0
        while True:
            left = 2 * index + 1
            right = 2 * index + 2
            largest = index 

            if left < len(self.heap) and self.heap[left] > self.heap[largest]:
                largest = left 
            
            if right < len(self.heap) and self.heap[right] > self.heap[largest]:
                largest = right 
            
            if largest != index:
                # Swap if one of the children is larger 
                self.heap[index], self.heap[largest] = self.heap[largest], self.heap[index]
                index = largest
            else:
                break 

class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        dict_a = {}
        for num in nums:
            if dict_a.get(num):
                dict_a[num] = dict_a.get(num) + 1
            else:
                dict_a[num] = 1

        max_heap = MaxHeap()
        for key, value in dict_a.items():
            max_heap.insert((value, key))
        
        result = []
        for i in range(k):
            result.append(max_heap.extract_max())

# ------------------------------------------------------------------------------------------
# Python heapq package 
import heapq 

class MedianFinder:
    def __init__(self): 
        # self.small : the smaller half
        # self.large : the bigger half
        self.small, self.large = [], [] # maxHeap, minHeap
    
    def addNum(self, num:int) -> None:
        if self.large and num > self.large[0]:
            heapq.heappush(self.large, num)
        else:
            heapq.heappush(self.small, -num)
        
        if len(self.small) > len(self.large) + 1:
            val = -1 * heapq.heappop(self.small)
            heapq.heappush(self.large, val)
        if len(self.large) > len(self.small) + 1:
            val = heapq.heappop(self.large)
            heapq.heappush(self.small, -1 * val)
    
        print('self.small is ->', self.small)
        print('self.large is ->', self.large)
        print('*******************************')

    def findMedian(self) -> float: 
        if len(self.small) > len(self.large):
            return -1 * self.small[0]
        elif len(self.large) > len(self.small):
            return self.large[0]
        return (-1 * self.small[0] + self.larg[0]) / 2.0


median_finder = MedianFinder()
median_finder.addNum(1)
median_finder.addNum(2)
median_finder.addNum(3)
median_finder.addNum(4)
median_finder.addNum(5)
```
