```python
from typing import Optional, List 
# 21. Merge Two Sorted Lists
"""
You are given the heads of two sorted linked lists list1 and list2.

Merge the two lists into one sorted list. The list should be made by splicing together the nodes of the first two lists.

Return the head of the merged linked list.
"""
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]: 
        # Create a dummy node to serve as the head of the merged list
        dummy = ListNode()
        current = dummy

        # Iterate through both lists until one of them is exhausted
        while list1 is not None and list2 is not None:
            if list1.val < list2.val:
                current.next = list1
                list1 = list1.next
            else:
                current.next = list2
                list2 = list2.next 
        
            # Move the current pointer to the last merged node
            current = current.next

        if list1 is not None:
            current.next = list1
        elif list2 is not None:
            current.next = list2
        
        return dummy.next
# -------------------------------------------------------------------------------
# 19. Remove Nth Node From End of List
"""
Given the head of a linked list, 
remove the nth node from the end of the list and return its head.
"""
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        # count list length first
        list_len = 0
        current = head
        while current:
            current = current.next
            list_len += 1
        t_index = list_len - n + 1
        
        cur = head 
        index = 0
        while cur:
            index += 1
            if index + 1 == t_index:
                next_n = cur.next
                next_next_n = next_n.next
                cur.next = next_next_n
                next_n.next = None
                break
            elif index == t_index:
                cur = cur.next
                head = cur
                return head
            cur = cur.next
        return head

# ---------------------------------------------------------
# 23. Merge k Sorted Lists
"""
You are given an array of k linked-lists lists, each linked-list is sorted in ascending order.

Merge all the linked-lists into one sorted linked-list and return it.
"""
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        linked_list = ListNode()
        head = linked_list
        compare = []
        while any(lists):
            for index, item in enumerate(lists):
                if item:
                    compare.append(item.val)
                    lists[index] = item.next
        s_compare = sorted(compare)
        linked_list = self.make_sorted_list(s_compare, linked_list)
        cur = head.next
        head = None 
        return cur

    def make_sorted_list(self, s_compare:list, linked_list: ListNode):
        cur = linked_list
        for val in s_compare:
            cur.next = ListNode(val = val, next = None)
            cur = cur.next
        return cur
    
# ------------------------------------------------------------------
# 143. Reorder List
"""
You are given the head of a singly linked-list. The list can be represented as:

L0 → L1 → … → Ln - 1 → Ln
Reorder the list to be on the following form:

L0 → Ln → L1 → Ln - 1 → L2 → Ln - 2 → …
You may not modify the values in the list's nodes. Only nodes themselves may be changed.
"""
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
import copy

class Solution:
    # First solution run in O(n2) times
    def reorderList(self, head: Optional[ListNode]) -> None:
        """
        Do not return anything, modify head in-place instead.
        """
        # count length of linked_list:
        linked_len = 0
        node = head
        while node:
            node = node.next
            linked_len += 1
        
        # get last node
        def get_last(node):
            while node.next:
                if not node.next.next:
                    last_node = node.next
                    node.next = None
                    return last_node
                node = node.next

        # Tow loop
        node = head
        count = 1
        
        if linked_len % 2 == 1:
            limit = (linked_len // 2) + 1
        else:
            limit = (linked_len // 2)
        
        # 1. Oppisite list
        while node and count < limit:
            next_head = node.next
            last_node = get_last(head)
            node.next = last_node
            last_node.next = next_head
            node = node.next.next
            count += 1
        return head

# ------------------------------------------------------------------
# 141. Linked List Cycle
"""
Given head, the head of a linked list, 
determine if the linked list has a cycle in it.
There is a cycle in a linked list if there is some node in the list that can be reached again by continuously following the next pointer. 
Internally, pos is used to denote the index of the node that tail's next pointer is connected to. 
Note that pos is not passed as a parameter.
Return true if there is a cycle in the linked list. Otherwise, return false.
"""
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        walked = {}
        node = head 
        while node:
            if not walked.get(node):
                walked[node] = 1
            else:
                return True 
            node = node.next
        return False 
```
